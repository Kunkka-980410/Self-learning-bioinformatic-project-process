```python
import pandas as pd
import re

# 假设df_af是您的AF样本DataFrame
print(f"AF样本矩阵形状: {df_af.shape}")
print("所有AF样本列名:")
print(df_af.columns.tolist())

# 创建分组信息
ctrl_group = []  # 对照组: I, II, I-II
deg_group = []   # 实验组: III, IV, V

# 为每个样本确定分组
sample_groups = {}
for col in df_af.columns:
    # 从列名中提取等级信息
    grade_match = re.search(r'_([IVI\-]+)$', col)
    if grade_match:
        grade = grade_match.group(1)
        
        if grade in ['I', 'II', 'I-II']:
            ctrl_group.append(col)
            sample_groups[col] = 'ctrl'
        else:
            deg_group.append(col)
            sample_groups[col] = 'deg'

# 显示分组结果
print(f"\n对照组 (I, II, I-II): {len(ctrl_group)} 个样本")
print(ctrl_group)
print(f"\n实验组 (III, IV, V): {len(deg_group)} 个样本")
print(deg_group)

# 创建分组DataFrame用于后续分析
group_df = pd.DataFrame({
    'sample': list(sample_groups.keys()),
    'group': list(sample_groups.values())
})
print("\n样本分组信息:")
print(group_df['group'].value_counts())

# 提取两组数据
df_ctrl = df_af[ctrl_group]  # 对照组数据
df_deg = df_af[deg_group]    # 实验组数据

print(f"\n对照组数据形状: {df_ctrl.shape}")
print(f"实验组数据形状: {df_deg.shape}")

# 为后续差异分析准备分组信息
# 方法1: 直接使用两组DataFrame
print("\n可以直接使用 df_ctrl 和 df_deg 进行差异分析")

# 方法2: 创建分组向量用于某些统计函数
group_vector = []
for col in df_af.columns:
    group_vector.append(0 if col in ctrl_group else 1)  # 0=ctrl, 1=deg

print(f"\n分组向量: {group_vector}")
```

```python
#✅ Python 模拟 limma 的推荐实现
#🔧 Step 1：构建表达矩阵和分组标签
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.multitest import multipletests


# 合并表达矩阵
df_combined = pd.concat([df_ctrl, df_deg], axis=1)

# 构建分组标签
group_labels = np.array([0]*df_ctrl.shape[1] + [1]*df_deg.shape[1])  # 0=Ctrl, 1=DEG
```

```python
#🔧 Step 2：使用 statsmodels 拟合线性模型
import statsmodels.api as sm

results = []

for gene in df_combined.index:
    y = df_combined.loc[gene].values.astype(float)
    X = sm.add_constant(group_labels)  # 添加截距项
    model = sm.OLS(y, X)
    try:
        fit = model.fit()
        log2fc = fit.params[1]  # 分组系数
        pval = fit.pvalues[1]
        results.append({'gene': gene, 'log2fc': log2fc, 'p_value': pval})
    except:
        results.append({'gene': gene, 'log2fc': np.nan, 'p_value': np.nan})
```

```python
#🔧 Step 3：FDR 校正 + 显著性筛选
deg_df = pd.DataFrame(results).set_index('gene')
deg_df['fdr'] = multipletests(deg_df['p_value'].fillna(1), method='fdr_bh')[1]
deg_df['significant'] = (deg_df['fdr'] < 0.05) & (abs(deg_df['log2fc']) > 1)
```

```python
#🔧 Step 4：输出结果
print(f"\n模拟 limma 差异分析结果:")
print(f"总基因数: {len(deg_df)}")
print(f"显著差异基因数: {deg_df['significant'].sum()}")

#deg_df.to_csv('af_limma_simulated_results.csv')
#print("结果已保存到 af_limma_simulated_results.csv")

top_degs = deg_df[deg_df['significant']].sort_values('log2fc', ascending=False)
print("\nTop 10 上调基因:")
print(top_degs.head(10)[['log2fc', 'fdr']])

print("\nTop 10 下调基因:")
print(top_degs.tail(10)[['log2fc', 'fdr']])
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager

# ✅ 设置中文字体（Windows系统推荐使用SimHei）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 支持中文
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# ✅ 准备数据
x = deg_df['log2fc']
y = -np.log10(deg_df['fdr'])

# ✅ 设置颜色
colors = []
for fc, fdr in zip(deg_df['log2fc'], deg_df['fdr']):
    if fdr < 0.05 and fc > 1:
        colors.append('red')   # 上调
    elif fdr < 0.05 and fc < -1:
        colors.append('blue')  # 下调
    else:
        colors.append('grey')  # 非显著

# ✅ 绘制火山图
plt.figure(figsize=(10, 8))
plt.scatter(x, y, c=colors, alpha=0.6, edgecolors='none')

# ✅ 添加阈值线
plt.axhline(-np.log10(0.05), color='black', linestyle='--', linewidth=1)
plt.axvline(1, color='black', linestyle='--', linewidth=1)
plt.axvline(-1, color='black', linestyle='--', linewidth=1)

# ✅ 标注前20个显著基因（按FDR升序）
top_genes = deg_df[deg_df['significant']].sort_values('fdr').head(20)
for gene in top_genes.index:
    plt.text(deg_df.loc[gene, 'log2fc'], 
             -np.log10(deg_df.loc[gene, 'fdr']), 
             gene, fontsize=8, ha='center', va='bottom')

# ✅ 图形美化
plt.title('AF分级差异表达火山图（Top 20基因标注）', fontsize=14)
plt.xlabel('log2倍数变化', fontsize=12)
plt.ylabel('-log10 FDR', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
