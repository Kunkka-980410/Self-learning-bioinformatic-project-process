```python
import pandas as pd
import re

# ÂÅáËÆædf_afÊòØÊÇ®ÁöÑAFÊ†∑Êú¨DataFrame
print(f"AFÊ†∑Êú¨Áü©ÈòµÂΩ¢Áä∂: {df_af.shape}")
print("ÊâÄÊúâAFÊ†∑Êú¨ÂàóÂêç:")
print(df_af.columns.tolist())

# ÂàõÂª∫ÂàÜÁªÑ‰ø°ÊÅØ
ctrl_group = []  # ÂØπÁÖßÁªÑ: I, II, I-II
deg_group = []   # ÂÆûÈ™åÁªÑ: III, IV, V

# ‰∏∫ÊØè‰∏™Ê†∑Êú¨Á°ÆÂÆöÂàÜÁªÑ
sample_groups = {}
for col in df_af.columns:
    # ‰ªéÂàóÂêç‰∏≠ÊèêÂèñÁ≠âÁ∫ß‰ø°ÊÅØ
    grade_match = re.search(r'_([IVI\-]+)$', col)
    if grade_match:
        grade = grade_match.group(1)
        
        if grade in ['I', 'II', 'I-II']:
            ctrl_group.append(col)
            sample_groups[col] = 'ctrl'
        else:
            deg_group.append(col)
            sample_groups[col] = 'deg'

# ÊòæÁ§∫ÂàÜÁªÑÁªìÊûú
print(f"\nÂØπÁÖßÁªÑ (I, II, I-II): {len(ctrl_group)} ‰∏™Ê†∑Êú¨")
print(ctrl_group)
print(f"\nÂÆûÈ™åÁªÑ (III, IV, V): {len(deg_group)} ‰∏™Ê†∑Êú¨")
print(deg_group)

# ÂàõÂª∫ÂàÜÁªÑDataFrameÁî®‰∫éÂêéÁª≠ÂàÜÊûê
group_df = pd.DataFrame({
    'sample': list(sample_groups.keys()),
    'group': list(sample_groups.values())
})
print("\nÊ†∑Êú¨ÂàÜÁªÑ‰ø°ÊÅØ:")
print(group_df['group'].value_counts())

# ÊèêÂèñ‰∏§ÁªÑÊï∞ÊçÆ
df_ctrl = df_af[ctrl_group]  # ÂØπÁÖßÁªÑÊï∞ÊçÆ
df_deg = df_af[deg_group]    # ÂÆûÈ™åÁªÑÊï∞ÊçÆ

print(f"\nÂØπÁÖßÁªÑÊï∞ÊçÆÂΩ¢Áä∂: {df_ctrl.shape}")
print(f"ÂÆûÈ™åÁªÑÊï∞ÊçÆÂΩ¢Áä∂: {df_deg.shape}")

# ‰∏∫ÂêéÁª≠Â∑ÆÂºÇÂàÜÊûêÂáÜÂ§áÂàÜÁªÑ‰ø°ÊÅØ
# ÊñπÊ≥ï1: Áõ¥Êé•‰ΩøÁî®‰∏§ÁªÑDataFrame
print("\nÂèØ‰ª•Áõ¥Êé•‰ΩøÁî® df_ctrl Âíå df_deg ËøõË°åÂ∑ÆÂºÇÂàÜÊûê")

# ÊñπÊ≥ï2: ÂàõÂª∫ÂàÜÁªÑÂêëÈáèÁî®‰∫éÊüê‰∫õÁªüËÆ°ÂáΩÊï∞
group_vector = []
for col in df_af.columns:
    group_vector.append(0 if col in ctrl_group else 1)  # 0=ctrl, 1=deg

print(f"\nÂàÜÁªÑÂêëÈáè: {group_vector}")
```

```python
#‚úÖ Python Ê®°Êãü limma ÁöÑÊé®ËçêÂÆûÁé∞
#üîß Step 1ÔºöÊûÑÂª∫Ë°®ËææÁü©ÈòµÂíåÂàÜÁªÑÊ†áÁ≠æ
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.multitest import multipletests


# ÂêàÂπ∂Ë°®ËææÁü©Èòµ
df_combined = pd.concat([df_ctrl, df_deg], axis=1)

# ÊûÑÂª∫ÂàÜÁªÑÊ†áÁ≠æ
group_labels = np.array([0]*df_ctrl.shape[1] + [1]*df_deg.shape[1])  # 0=Ctrl, 1=DEG
```

```python
#üîß Step 2Ôºö‰ΩøÁî® statsmodels ÊãüÂêàÁ∫øÊÄßÊ®°Âûã
import statsmodels.api as sm

results = []

for gene in df_combined.index:
    y = df_combined.loc[gene].values.astype(float)
    X = sm.add_constant(group_labels)  # Ê∑ªÂä†Êà™Ë∑ùÈ°π
    model = sm.OLS(y, X)
    try:
        fit = model.fit()
        log2fc = fit.params[1]  # ÂàÜÁªÑÁ≥ªÊï∞
        pval = fit.pvalues[1]
        results.append({'gene': gene, 'log2fc': log2fc, 'p_value': pval})
    except:
        results.append({'gene': gene, 'log2fc': np.nan, 'p_value': np.nan})
```

```python
#üîß Step 3ÔºöFDR Ê†°Ê≠£ + ÊòæËëóÊÄßÁ≠õÈÄâ
deg_df = pd.DataFrame(results).set_index('gene')
deg_df['fdr'] = multipletests(deg_df['p_value'].fillna(1), method='fdr_bh')[1]
deg_df['significant'] = (deg_df['fdr'] < 0.05) & (abs(deg_df['log2fc']) > 1)
```

```python
#üîß Step 4ÔºöËæìÂá∫ÁªìÊûú
print(f"\nÊ®°Êãü limma Â∑ÆÂºÇÂàÜÊûêÁªìÊûú:")
print(f"ÊÄªÂü∫Âõ†Êï∞: {len(deg_df)}")
print(f"ÊòæËëóÂ∑ÆÂºÇÂü∫Âõ†Êï∞: {deg_df['significant'].sum()}")

#deg_df.to_csv('af_limma_simulated_results.csv')
#print("ÁªìÊûúÂ∑≤‰øùÂ≠òÂà∞ af_limma_simulated_results.csv")

top_degs = deg_df[deg_df['significant']].sort_values('log2fc', ascending=False)
print("\nTop 10 ‰∏äË∞ÉÂü∫Âõ†:")
print(top_degs.head(10)[['log2fc', 'fdr']])

print("\nTop 10 ‰∏ãË∞ÉÂü∫Âõ†:")
print(top_degs.tail(10)[['log2fc', 'fdr']])
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager

# ‚úÖ ËÆæÁΩÆ‰∏≠ÊñáÂ≠ó‰ΩìÔºàWindowsÁ≥ªÁªüÊé®Ëçê‰ΩøÁî®SimHeiÔºâ
plt.rcParams['font.sans-serif'] = ['SimHei']  # ÊîØÊåÅ‰∏≠Êñá
plt.rcParams['axes.unicode_minus'] = False    # Ê≠£Â∏∏ÊòæÁ§∫Ë¥üÂè∑

# ‚úÖ ÂáÜÂ§áÊï∞ÊçÆ
x = deg_df['log2fc']
y = -np.log10(deg_df['fdr'])

# ‚úÖ ËÆæÁΩÆÈ¢úËâ≤
colors = []
for fc, fdr in zip(deg_df['log2fc'], deg_df['fdr']):
    if fdr < 0.05 and fc > 1:
        colors.append('red')   # ‰∏äË∞É
    elif fdr < 0.05 and fc < -1:
        colors.append('blue')  # ‰∏ãË∞É
    else:
        colors.append('grey')  # ÈùûÊòæËëó

# ‚úÖ ÁªòÂà∂ÁÅ´Â±±Âõæ
plt.figure(figsize=(10, 8))
plt.scatter(x, y, c=colors, alpha=0.6, edgecolors='none')

# ‚úÖ Ê∑ªÂä†ÈòàÂÄºÁ∫ø
plt.axhline(-np.log10(0.05), color='black', linestyle='--', linewidth=1)
plt.axvline(1, color='black', linestyle='--', linewidth=1)
plt.axvline(-1, color='black', linestyle='--', linewidth=1)

# ‚úÖ Ê†áÊ≥®Ââç20‰∏™ÊòæËëóÂü∫Âõ†ÔºàÊåâFDRÂçáÂ∫èÔºâ
top_genes = deg_df[deg_df['significant']].sort_values('fdr').head(20)
for gene in top_genes.index:
    plt.text(deg_df.loc[gene, 'log2fc'], 
             -np.log10(deg_df.loc[gene, 'fdr']), 
             gene, fontsize=8, ha='center', va='bottom')

# ‚úÖ ÂõæÂΩ¢ÁæéÂåñ
plt.title('AFÂàÜÁ∫ßÂ∑ÆÂºÇË°®ËææÁÅ´Â±±ÂõæÔºàTop 20Âü∫Âõ†Ê†áÊ≥®Ôºâ', fontsize=14)
plt.xlabel('log2ÂÄçÊï∞ÂèòÂåñ', fontsize=12)
plt.ylabel('-log10 FDR', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
